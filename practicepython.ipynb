```python
try:
    rules = {
    "rock": "scissors",
    "scissors": "paper",
    "paper": "rock"
    }
    def player_choice(player):
        txt = f"player {player} , type your choice. (paper, rock or scissors): "
        choice = input(txt)
        while choice.lower() not in rules:
            print("please enter the correct characters.\n try again!")
            choice = input(txt)
        return choice  

    def game(player1_choice,player2_choice):
        if player1_choice.lower() == player2_choice.lower():
            return("Same choices!")
        for k in rules:
            if player1_choice.lower() == k and player2_choice.lower() == rules[k]:
                return("player one won the game")
            return("player two won the game")
            
    while True: 
        player_one = player_choice("one")
        player_two = player_choice("two")
        result = game(player_one, player_two)
        print(result)
        play_again = input("do you wanna play again?: ")
        
        if play_again.lower() in ("n","no"): 
            print("game is end")
            break
    
except Exception as e:
    print("Something went wrong.")
    print(e)
```

# Exercise 18


Randomly generate a 4-digit number. Ask the user to guess a 4-digit number. For every digit that the user guessed correctly in the correct place, they have a “cow”. For every digit the user guessed correctly in the wrong place is a “bull.” Every time the user makes a guess, tell them how many “cows” and “bulls” they have. Once the user guesses the correct number, the game is over. Keep track of the number of guesses the user makes throughout the game and tell the user at the end.

```python
import random

class CowAndBull:

    def user_input():

        while True:
            user_number = input("write your guess (or type exit): ")

            if user_number.lower() == "exit":
                return "exit"

            if len(user_number) != 4 or not user_number.isdigit():
                print("Your guess must contain exactly 4 digits. Try again.")
                continue

            return [int(digit) for digit in user_number]

    def num_generator():
        return random.sample(range(0, 10), 4)

    def play():

        guesses = 0
        cpu = CowAndBull.num_generator()

        while True:

            user = CowAndBull.user_input()

            if user == "exit":
                print("Game exited. The secret number was:", cpu)
                break

            guesses += 1

            cow_counter = 0
            bull_counter = 0

            for i in range(4):

                if user[i] == cpu[i]:
                    cow_counter += 1

                elif user[i] in cpu:
                    bull_counter += 1

            print(f"Cows: {cow_counter}")
            print(f"Bulls: {bull_counter}\n")

            if user == cpu:
                print("Congratulations! You guessed the number!")
                print(f"You took {guesses} guesses.")
                break


CowAndBull.play()
```
